```python
# Set notebook to display figures inline and import dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
import os
```

```python
# Set path for csv data
data_path = os.path.join(".","Resources","Pyber_ride_data.csv")

# Create data frame pyber_ride_df from csv file
pyber_ride_df = pd.read_csv(data_path)

# Set tick locations to dislay each month
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Create a bar Chart from pyber_ride_df
stdev = pyber_ride_df["Avg. Fare ($USD)"].std()
print(stdev)
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", yerr = stdev, kind = "bar", capsize = 3,color = "skyblue")
plt.xticks(rotation = "horizontal")
```
