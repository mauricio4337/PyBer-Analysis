```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import statistics
import pandas as pd
```

```python
np.arange(0,22,step=2)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create plot and labels and title
plt.plot(x_axis, y_axis,marker='d',color='green',linewidth=1,label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.xlim(0,8)
plt.title('PyBer Fare by Month')
plt.legend()
# Add a grid
plt.grid()
```

```python
# Create a bar chart with the same data, labels and title
plt.barh(x_axis,y_axis,label = 'Boston',color='green')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.xlim(0,45)
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color='magenta',label='Boston')
plt.gca().invert_yaxis()
plt.legend()
plt.xticks(np.arange(0,46,step=5))
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
```

```python
# Create a scatter plot using plot method
plt.scatter(x_axis, y_axis,s=100*[10,200,300,420,55,600,70,25,45,500,20,100],c='skyblue',
            edgecolor='black',alpha=0.8,
            label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.legend()
plt.title('PyBer Fare by Month')
```

```python
plt.subplots(figsize=(8, 8))
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, colors=colors,autopct='%.1f%%')
plt.show()

```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)

plt.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```
